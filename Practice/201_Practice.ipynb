```python
# create a numpy array with 4 rows and 5 columns
# values between 1 and 20, using arange and reshape
np.arange(1, 21).reshape(4, 5)
```

```python
# create a andas data from it
pd.DataFrame(np.arange(1, 21).reshape(4, 5))
```

```python
# assign column names as c1, c2, etc
pd.DataFrame(np.arange(1, 21).reshape(4, 5), columns=['c1', 'c2', 'c3', 'c4', 'c5'])

# another way
pd.DataFrame(np.arange(1, 21).reshape(4, 5), columns='c1 c2 c3 c4 c5'.split())
```

```python
# assign row lables or indexes as r1, r2, etc
pd.DataFrame(np.arange(1, 21).reshape(4, 5), columns='c1 c2 c3 c4 c5'.split(), index='r1 r2 r3 r4'.split())
```

```python
# assign it to variable
df = pd.DataFrame(np.arange(1, 21).reshape(4, 5), columns='c1 c2 c3 c4 c5'.split(), index='r1 r2 r3 r4'.split())
df
```

```python
# use sample function to get two random samples of the dataframe
df.sample(2)
```

```python
# use loc to ge the following
# 7, 8, 9
# 12, 13, 14
df.loc['r2':'r4','c2':'c4']
```
