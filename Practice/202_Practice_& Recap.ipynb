- as we have null values in two different columns, namely Age and Major

- while one is a categorical column, the other one stores numerical data, for each one we must appky appropriate methods

- for age, which is a numerical column, and the age of students must be close to other students, so we use average to fill this null value

```python
students_df.Age.mean()
# average age of students
```

```python
students_df.Age.fillna(students_df.Age.mean(), inplace = True)
```

```python
students_df
```

- for numerical columns, mean and median techniques are used to fill the null values
    
    - use mean if there are no outliers

    - use median if there are outliers.

- for categorical columns

    - use mode = most repeated or most frequent value if:
    
        - we have one value repeated too many times and the other categories repeated less

    - otherwise use bfill and ffill techniques that randomly fill the null values.

        - while filling the null values in categorical column, best strategy would be to find similarities in rows.

```python
students_df['Major'].mode()
# most frequent value in the Major column
# returns a series
```

```python
students_df['Major'].fillna(students_df['Major'].mode()[0])
# fills the null values with the most frequent value in the Major column
# we use [0] because mode() returns a series, and we need the first value of that series
```

```python
students_df
```

```python
students_df['Major'].fillna(method = 'bfill')
# fills the null values with the next valid value in the Major column
```

```python
students_df['Major'].fillna(method = 'ffill')
# fills the null values with the previous valid value in the Major column
```
