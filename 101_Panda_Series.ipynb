## Creating a Pandas Series

- Panda series is 1D data structure
- Objects can hold many type of data, including objects, flots, strings, and intigers
- You can create a series by calling pandas.Series()
- A list, numpy array, and dictionary can be converted to a pandas array
- Axis labels are collectively called index
- Difference between pandas array and numpy array is that they keep values with INDEXES

## Using Scalar Value

## Indexing and Slicing with Panda Series

```python
series6 = pd.Series([1,2 ,3 ,4, 5], index=['asus', 'dell', 'lenovo','samsung', 'apple'])
series6
```

```python
series6[0]
# will return index 0, which is value of index 'asus', wich is 1
# default indexes are always avaialble for use
```

```python
# ignore the warning from Python
import warnings
warnings.filterwarnings('ignore')
```

```python
# warnings are ignored
series6[3]
```

```python
# slicing
series6[2:5]
```

```python
series6['asus':'samsung']
# when you use the labled indexes, the ending point will be included
# when you use the default indexes, lik1 0, 1, 2, the ending point will not be included
```

```python
series6['dell':]
# it will give the values starting from 'dell' till hte end
```

```python
series6['asus':'apple':2]
```

```python
series6['asus':'samsung':-1]
# will return empty series
```

```python
series6['samsung':'asus':-1]
```

```python
series6[::-2]
# reverse the series
```

```python
series6[['asus', 'lenovo', 'apple']]
# we can select the data that we would like to see
# we use double square bracket for it []
```

```python
series6['asus', 'lenovo', 'apple']
# will give an error, we need to use [] to get the multiple specific indexes
```

## <font color='red'> <b>Basic Attributes & Methods of Series</b><font color='black'>

**Series.values:** Allows the values of the series to be returned as a Numpy array.

**Series.index:** Allows the indexes of the series to be returned.

**Series.dtype:** Returns the data type of the series.

**Series.size:** Returns the number of elements of the series.

**Series.shape:** Returns the size of the series.

**Series.ndim:** Returns the dimension number of the series.

**Series.head():** Returns the first n elements of the series.

**Series.tail():** Returns the last n elements of the series.

**Series.sample():** Returns n random elements of the series.

**Series.describe():** Returns the statistical summary of the series.

**Series.sort_index:** Sorts the series according to their index.

**Series.sort_values():** Sorts the series by its values.

**Series.isnull():** Checks whether each element of the series is null (None).

**Series.fillna():** Fills null values with a specified value.
 
**Series.dropna():** Drops null values from the series.

**Series.isin:** Checks whether the elements in the series are included in the values.
