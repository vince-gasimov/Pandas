- when there is a nan value in a numeric column how to fill in ?
    - Mean -> aritmetic average might be sensible most of the time , but if your data set is small and you have outliers,
    - it might be more sensible to use median instead of mean. because median is not effected from outliers. 
    - According to the data set that you have , you will need to decide what to use. 

- differences between mean and median
    - mean is average, it represents the average number from dataset
        - for example [2, 3, 3, 5, 6, 9, 8, 11, 12], mean is 6.5
        - 6.5 represent every data set, it close number

    - median is number right in the middle
        - if you have data set [2, 3, 4, 5, 6, 9, 8, 1100, 1200], mean is 259
        - 259 does not represent neither small nor big numbers
        - it make sense to take median, number in middle, 6
        - 6 represent majority

### <font color='yellow'> <b>Filling Categorical Variables</b><font color='orange'>

- whenever we have categorical data to fill in, we use mode technique
    - with mode technique, we fill in nan values with most used catergirical data
    
- gender -> 
    - M -> Male 
    - F -> Female 

- status 
    - FT -> Full time 
    - PT -> Part time 

- dept -> department 
    - DS -> Data science 
    - FS -> Full stack 
    - AWS ->

- mode  -> repeats at most 
    - 3 fs 
    - 1 ds 
    - 7 aws
    - mode -> aws

```python
df
```

```python
df.gender.value_counts()
# count unique data
```

```python
df.dept.value_counts()
```

```python
# out in most used data order
df.dept.mode()
# it will return series
```

```python
df.dept.mode()[0]
# the most repeated value among them
# we can index ths series and get the first data
```

```python
df.dept
```

```python
df.dept.fillna(df.dept.mode()[0])
# change is not perminant
```

```python
df.dept
```

```python
df.dept.fillna(method = 'bfill')
# back fill --> it fills in the previous value with the next one
```

```python
df.dept
```

```python
df.dept.fillna(method = 'ffill', inplace = True)
# forward fill --> it fills in the next value with the previous one
```

```python
df.dept
```

### <font color='yellow'> <b>Filling by condition & group</b><font color='black'>

```python
df
```

```python
# first let's find the min, max and average salary of full time and part time employee
df.groupby('status')['salary'].agg([min, max, np.mean])
```

- there is a clear salary difference between part time PT and full time FT employee
- minimum salary of FT employee is 120.0
- maximum salary of PT employee is 61
- we cannot just fill out avaerage number of salary column

```python
df.loc[df.salary > 90, 'status']
```

```python
df.loc[df.salary > 90, 'status'].mode()[0]
# people who earn more than 90, if mostly FT employee
```

```python
df.loc[df.salary <= 90, 'status']
# who are the low earners
```

```python
df.loc[df.salary <= 90, 'status'].mode()[0]
# almost all are part time employee
```

```python
# go to less earners, and if there is a null value, fill it with the mode (most used) of this category.

# first way
df.loc[df.salary <= 90, 'status'].fillna(df.loc[df.salary <= 90, 'status'].mode()[0], inplace = True)
```

```python
# second way
# you can assigne the category of low eraners to variable and use it
categorry_of_low_earners = df.loc[df.salary <= 90, 'status'].mode()[0]
categorry_of_low_earners
```

```python
df
# even if i used inplace = True parameter, the status hasn't changed permanently.
```

```python
df.loc[df.salary <= 90, 'status']
# and the reason is, while filling the null values, we made this selection, or filtering, so python is filling 
# the null values, in the following selected series, not in the original dataframe, in these type of cases
# we must use reassignment technique.
```

```python
# first way of perminantly assigning
df.loc[df.salary <= 90, 'status'] = df.loc[df.salary <= 90, 'status'].mode()[0]
# the assignment technique, where we select the series, and assign PT value to the entire series.
```

```python
# second way of perminantly assigning
df.loc[df.salary <= 90, 'status'] = categorry_of_low_earners
```

```python
df
```

```python
df.loc[df.salary > 90, 'status']
```

```python
categorry_of_high_earners = df.loc[df.salary > 90, 'status'].mode()[0]
categorry_of_high_earners
```

```python
# df.loc[df.salary > 90, 'status'].fillna(df.loc[df.salary > 90, 'status'].mode()[0], inplace = True)
df.loc[df.salary > 90, 'status'].fillna(categorry_of_high_earners, inplace = True)
```

```python
df
```

```python
# df.loc[df.salary <= 90, 'status'] = df.loc[df.salary <= 90, 'status'].mode()[0]
df.loc[df.salary > 90, 'status'] = categorry_of_high_earners
```
